```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
kilo=[20,60,80,85,86,87,70,90,95,90]
```

```python
numpydizi1=np.linspace(0,10,20)
```

```python
numpydizi1
```

```python
numpydizi2= numpydizi1 **3
```

```python
numpydizi2
```

```python
plt.plot(numpydizi1,numpydizi2,"b*-")
```

```python
plt.subplot(1,2,1) #1 sıra olcak ,2 grafik olcak, ilk grafik işlemi
plt.plot(numpydizi1,numpydizi2,"r*-")
plt.subplot(1,2,2) #1 sıra olcak ,2 grafik olcak, ikinci grafik işlemi
plt.plot(numpydizi2,numpydizi1,"g--")
```

```python
figure=plt.figure()
figureAxes=figure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpydizi1,numpydizi2,"g")
figureAxes.set_xlabel("x ekseni veri ismi")
figureAxes.set_ylabel("y ekseni veri ismi")
figureAxes.set_title("grafik başlığı")
```

```python
figure2=plt.figure()
eksen1=figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2=figure2.add_axes([0.2,0.5,0.3,0.3])
eksen1.plot(numpydizi1,numpydizi2,"g")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("ana grafik başlık")


eksen2.plot(numpydizi2,numpydizi1,"b")
eksen2.set_xlabel("x ekseni küçük")
eksen2.set_ylabel("y ekseni küçük ")
eksen2.set_title("küçük başlık")
```

```python
(benimfigure,benimeksen)=plt.subplots(nrows=1,ncols=2)
type(benimeksen)
for eksen in benimeksen:
    eksen.plot(numpydizi1,numpydizi2,"g")
    eksen.set_xlabel("x ekseni")
plt.tight_layout()
    
```

```python
yeniFigur=plt.figure(figsize=(4,4))

yeniEksen=yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpydizi1,numpydizi1**2,label="numpydizi ** 2")
yeniEksen.plot(numpydizi1,numpydizi1**3,label="numpydizi ** 3")
yeniEksen.legend(loc=6)
```
