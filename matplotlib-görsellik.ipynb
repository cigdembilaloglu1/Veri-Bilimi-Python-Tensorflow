```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpydizi=np.linspace(0,10,20)
```

```python
numpydizi
```

```python
numpydizi2=numpydizi **2
```

```python
(benimFigur,benimEksen)=plt.subplots()
benimEksen.plot(numpydizi,numpydizi2,color="#35294D")
benimEksen.plot(numpydizi2,numpydizi,color="#BD3F93")
```

```python
(yeniFigur,yeniEksen)=plt.subplots()
yeniEksen.plot(numpydizi,numpydizi +2,color="blue",linewidth=3.0,linestyle="-.")
yeniEksen.plot(numpydizi,numpydizi +3,color="yellow",linewidth=4.0,linestyle=":")
yeniEksen.plot(numpydizi,numpydizi +4,color="#3AE81F",linestyle="--")
yeniEksen.plot(numpydizi,numpydizi +5,color="black",linestyle="--",marker="o",markersize=8,markerfacecolor="red")
yeniEksen.plot(numpydizi,numpydizi +6,color="red",linestyle="-.")
```
